```python
columns_by_domain = [
    col for col in eda_df.columns 
    if 'Examiner' in col 
    or ('SS' in col and 'CSS' not in col)
]
columns_by_domain
```

```python
eda_new_df = eda_df[columns_by_domain]
eda_new_df
```

```python
column_pairs = list(zip(eda_new_df.columns[::2], eda_new_df.columns[1::2]))
column_pairs
```

```python
examiner_dict = {}

for examiner, score in column_pairs:
    examiner_dict[f"{examiner}_{score}"]=eda_new_df.groupby(examiner)[score].mean().reset_index()

print(examiner_dict)
```

```python
examiner_dict.keys()
```

```python
examiner_dict['Adaptive-Self Care Examiner_Adaptive-Self Care SS']
```

```python
dfs=[]

for df in examiner_dict.values():
    temp = df.rename(columns={df.columns[0]:'examiner'})
    dfs.append(temp)

print(dfs)
```

```python
merged=dfs[0]

for df in dfs[1:]:
    merged=pd.merge(
        left=merged,
        right=df,
        on='examiner',
        how='outer')
merged
```

```python
merged['overall_avg'] = merged.mean(axis=1, numeric_only=True)
merged['overall_avg']
```

```python
merged[['examiner', 'overall_avg']].sort_values(by='overall_avg', ascending=False)
```

```python
merged.boxplot(
    column='overall_avg',
    by='examiner',
    vert=False,
    grid=False,
    figsize=(10, 30)
)
```

```python
location_mappings = []

for col in examiner_cols:
    subset = eda_df[[col, 'Location - Sub Level 1']].dropna()
    subset.columns = ['examiner', 'region']
    location_mappings.append(subset)
```
